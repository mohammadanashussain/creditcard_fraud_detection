```python
#separating the data for analysis
legit = credit_card_data[credit_card_data['Class'] == 0]
print("Number of valid Transaction: ",legit.shape)
fraud = credit_card_data[credit_card_data['Class'] == 1]
print("Number of fraud Transaction: ",fraud.shape)
```

Under-Sampling

Build a samle dataset containing similar distribution of normal transaction and fraudlent transaction

Number of fraudlent transacton 492

```python
legit_sample = legit.sample(n = 492)
```

concatenating two dataframes

```python
new_dataset = pd.concat([legit_sample,fraud],axis = 0)
new_dataset.head()
```

```python
new_dataset.tail()
```

```python
new_dataset['Class'].value_counts()
```

```python
new_dataset.groupby('Class').mean()
```

splitting the data into feature and targets

```python
X = new_dataset.drop(columns ='Class' , axis = 1)
Y = new_dataset['Class']
```

```python
print(X)
```

```python
print(Y)
```

To split the data into training data & testing data

```python
X_train, X_test ,Y_train , Y_test = train_test_split(X, Y, test_size = 0.2, stratify = Y, random_state= 2)
```

```python
print(X.shape,X_train.shape, X_test.shape)
```

```python
print(Y.shape,Y_train.shape, Y_test.shape)
```

Model Training

Logistic Regression - Generally used for binary of classification of data

```python
model = LogisticRegression()
```

```python
# training the Logistic Regression Model with Training Data
model.fit(X_train, Y_train)
```

Model Evaluation

Accuracy Score

```python
#accuraccy on training data
X_train_prediction =model.predict(X_train)
training_data_accuracy = accuracy_score(X_train_prediction, Y_train)
```

```python
print('Accuracy on Training data : ',training_data_accuracy)
```

```python
# accuracy on test data
X_test_prediction = model.predict(X_test)
test_data_accuracy = accuracy_score(X_test_prediction, Y_test)
```

```python
print('Accuracy score on Test data : ',test_data_accuracy)
```

```python
# if the accuracy score of training data ad test data is not similar or nearly same its means our model is over-fitted and under-fitted
#accuracy score of training data >>>>>>>> accuracy score of test data                                    Over-fitting
#accuuracy score of training data <<<<<<< accuracy score of test data                                    Under_fitting
```
